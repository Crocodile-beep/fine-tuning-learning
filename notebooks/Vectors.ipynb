```python
import numpy as np
import matplotlib.pyplot as plt

cat_vector = np.array([5, 3])  # Вектор "кот"
house_vector = np.array([8, 6])  # Вектор "дом" — уют, размер
combined = np.add(cat_vector, house_vector)  # Сложение
print("Кот + Дом:", combined)  # Комбинирует кота и дом — уютный кот в доме

big_house = house_vector * 1.5  # Скалярное умножение
print("Большой дом:", big_house)  # Усиливает уют и размер — большой дом

dot_product = np.dot(cat_vector, house_vector)  # Дот-продукт
print("Похожесть кота и дома:", dot_product)  # Похожесть кота и дома — низкая, разные сущности

garden_vector = np.array([3, 7])  # Вектор "сад"
combined_garden = np.add(house_vector, garden_vector)
dot_garden = np.dot(house_vector, garden_vector)
print("Дом + Сад:", combined_garden)  # Сложение дома и сада
print("Похожесть дома и сада:", dot_garden)  # Умеренная похожесть (места)
```
